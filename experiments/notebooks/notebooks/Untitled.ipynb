```python
import pandas as pd

# Cargar dataset base v21
df = pd.read_csv("../data/clean/dataset_base_v21.csv")

print("Columnas disponibles:")
print(df.columns.tolist())

df.head()
```

```python
if "token_density" in df.columns:
    print("✔ token_density encontrado")
else:
    print("❌ token_density NO encontrado — necesitas calcularlo antes")
```

```python
print("Distribución token_density por clase:")
display(df.groupby("label")["token_density"].describe())

print("\nFalsos positivos potenciales (legítimos con density > 0.5):")
display(
    df[(df["label"] == 0) & (df["token_density"] > 0.5)]
    [["url", "token_density"]]
    .head(10)
)

print("\nFalsos negativos potenciales (phishing con density < 0.2):")
display(
    df[(df["label"] == 1) & (df["token_density"] < 0.2)]
    [["url", "token_density"]]
    .head(10)
)
```
