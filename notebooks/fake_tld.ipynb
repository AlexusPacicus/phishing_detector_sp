```python
import pandas as pd

df = pd.read_csv("../data/clean/dataset_base_v21.csv")

print("### Distribución por clase ###")
display(df.groupby("label")["fake_tld_in_subdomain_or_path"].describe())

print("\n### FP potenciales ###")
display(df[(df["label"] == 0) & (df["fake_tld_in_subdomain_or_path"] > 0)][["url","fake_tld_in_subdomain_or_path"]].head(20))

print("\n### FN potenciales ###")
display(df[(df["label"] == 1) & (df["fake_tld_in_subdomain_or_path"] == 0)][["url","fake_tld_in_subdomain_or_path"]].head(20))

print("\n### Value counts ###")
print(df["fake_tld_in_subdomain_or_path"].value_counts())
```
