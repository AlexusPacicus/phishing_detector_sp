```python
import pandas as pd

df = pd.read_csv("../data/clean/dataset_base_v21.csv")

print("### Distribución por clase ###")
display(df.groupby("label")["param_count_boost"].describe())

print("\n### FP potenciales (legítimos con valor > 0) ###")
display(df[(df["label"] == 0) & (df["param_count_boost"] > 0)][["url","param_count_boost"]].head(15))

print("\n### FN potenciales (phishing con valor = 0) ###")
display(df[(df["label"] == 1) & (df["param_count_boost"] == 0)][["url","param_count_boost"]].head(15))

print("\n### Value counts ###")
print(df["param_count_boost"].value_counts())
```
