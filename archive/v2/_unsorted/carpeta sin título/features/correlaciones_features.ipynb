```python
import pandas as pd, numpy as np
import seaborn as sns, matplotlib.pyplot as plt

df = pd.read_csv("../dataset/eval_set_inclusion_v2.csv")

# ==== columnas numéricas de features (ajusta si tienes más/menos) ====
feature_cols = [
    "domain_length","domain_entropy","domain_complexity","tld_risk_weight",
    "host_entropy","param_count_boost","token_density","trusted_path_token",
    "trusted_path_penalty","suspicious_path_token","whitelist_strength",
    "protocol_context_penalty","free_hosting_score","free_hosting_weighted",
    "oauth_like_relief"
]
X = df[feature_cols].copy()

# Sanidad: numérico y sin NaN
assert X.isna().sum().sum() == 0, "Hay NaN en features"
for c in X.columns: X[c] = X[c].astype(float)

corr_p = X.corr("pearson")
corr_s = X.corr("spearman")

# Pares |r|>=0.8 (Pearson)
pairs = (corr_p.where(np.triu(np.ones(corr_p.shape),k=1).astype(bool))
         .stack().reset_index())
pairs.columns = ["A","B","pearson_r"]
pairs["|r|"]=pairs["pearson_r"].abs()
pairs_high = pairs.loc[pairs["|r|"]>=0.8].sort_values("|r|",ascending=False)
display(pairs_high)

# Heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr_p, annot=False, cmap="coolwarm", square=True)
plt.title("Pearson correlations – Features v2")
plt.tight_layout(); plt.savefig("../docs/img/features_v2_correlation_heatmap.png", dpi=220)
plt.show()
```
