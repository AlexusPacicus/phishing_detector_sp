```python
import pandas as pd
from pathlib import Path

# Detectar raíz del proyecto correctamente
def detect_repo_root():
    current = Path.cwd()
    for parent in [current] + list(current.parents):
        if (parent / ".git").exists() or (parent / "scripts").exists():
            return parent
    raise RuntimeError("No se detectó la raíz del proyecto")

REPO_ROOT = detect_repo_root()
DATA_INTERIM = REPO_ROOT / "data" / "interim" / "phishing"

# Ruta corregida
path = DATA_INTERIM / "primer_phishingrepo.csv"

# Cargar CSV
df = pd.read_csv(path)

# Añadir columna manual si no está
if "label_manual" not in df.columns:
    df["label_manual"] = None

# Confirmar
print(f"✅ Cargadas {len(df)} URLs para curación manual")
df[["url", "score_total", "signals_detected", "label_manual"]].head(10)
```

```python
# === Exportar curación manual para Excel ===

# Filtrar columnas útiles
df_excel = df[["url", "score_total", "signals_detected"]].copy()
df_excel["label_manual"] = None  # columna a completar manualmente

# Ruta de salida
export_path = DATA_INTERIM / "primer_phishingrepo_para_excel.csv"

# Guardar como CSV (editable en Excel sin problemas)
df_excel.to_csv(export_path, index=False, encoding="utf-8")

print(f"✅ Archivo preparado para curación en Excel: {export_path}")
print("✍️ Puedes etiquetar directamente en la columna 'label_manual'")
```
