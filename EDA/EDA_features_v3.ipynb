```python
# === Validaciones de dominio para FEATURES_V3 ===

print("Validación domain_complexity en rango [0,1]:",
      df["domain_complexity"].between(0,1).all())

print("Valores únicos domain_whitelist:", df["domain_whitelist"].unique())
print("Validación domain_whitelist ∈ {0,1}:",
      set(df["domain_whitelist"].unique()) <= {0,1})

print("Valores únicos trusted_token_context:", df["trusted_token_context"].unique())
print("Validación trusted_token_context ∈ {-1,0,1}:",
      set(df["trusted_token_context"].unique()) <= {-1,0,1})

print("Validación brand_in_path ∈ {0,1}:",
      set(df["brand_in_path"].unique()) <= {0,1})

print("Validación brand_match_flag ∈ {0,1}:",
      set(df["brand_match_flag"].unique()) <= {0,1})
```
