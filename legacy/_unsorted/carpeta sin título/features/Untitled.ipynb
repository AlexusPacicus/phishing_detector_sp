```python
# ==============================================================
# ğŸ”— Fusionar dataset_prototipo (categoria, matched_target)
#    con features_prototipo (features numÃ©ricas)
# ==============================================================

import pandas as pd
from pathlib import Path

# --------------------------------------------------------------
# 1ï¸âƒ£ Cargar datasets
# --------------------------------------------------------------
path_meta = Path("../dataset/dataset_prototipo.csv")      # contiene categoria y matched_target
path_feat = Path("features_prototipo.csv")     # contiene las features

df_meta = pd.read_csv(path_meta)
df_feat = pd.read_csv(path_feat)

print(f"âœ… dataset_prototipo.csv: {df_meta.shape}")
print(f"âœ… features_prototipo.csv: {df_feat.shape}")

# --------------------------------------------------------------
# 2ï¸âƒ£ Normalizar columnas
# --------------------------------------------------------------
df_meta.columns = df_meta.columns.str.strip().str.lower()
df_feat.columns = df_feat.columns.str.strip().str.lower()

# Detectar columna comÃºn
join_col = None
for col in ["url", "domain"]:
    if col in df_meta.columns and col in df_feat.columns:
        join_col = col
        break

if join_col is None:
    raise ValueError("âŒ No se encontrÃ³ columna 'url' ni 'domain' comÃºn para hacer merge.")

print(f"ğŸ”— Combinando datasets usando columna '{join_col}'")

# --------------------------------------------------------------
# 3ï¸âƒ£ Seleccionar columnas relevantes del meta
# --------------------------------------------------------------
cols_to_add = [join_col, "categoria", "matched_target"]
df_meta = df_meta[cols_to_add]

# --------------------------------------------------------------
# 4ï¸âƒ£ Fusionar ambos datasets
# --------------------------------------------------------------
merged = df_feat.merge(df_meta, on=join_col, how="left")

print(f"âœ… Dataset fusionado: {merged.shape}")
print(f"â„¹ï¸ Coincidencias encontradas: {merged['categoria'].notna().sum()} de {len(merged)} filas")

# --------------------------------------------------------------
# 5ï¸âƒ£ Guardar dataset enriquecido
# --------------------------------------------------------------
output_path = Path("features_prototipo_con_sector_entidad.csv")
merged.to_csv(output_path, index=False)
print(f"ğŸ’¾ Guardado en: {output_path.resolve()}")
```
