```python
import pandas as pd

# 1️⃣ Cargar el archivo descargado
df = pd.read_csv("../../tranco.csv", names=["rank", "domain"])

# 2️⃣ Filtrar solo los dominios .es
df_es = df[df["domain"].str.endswith(".es")].reset_index(drop=True)

# 3️⃣ Mantener los 200 primeros (suficiente para fuzzy match)
df_es_top = df_es.head(200)

# 4️⃣ Guardar en tu estructura del proyecto
df_es_top.to_csv("../../docs/spanish_domains.csv", index=False)

print(f"✅ Generado data/spanish_domains.csv con {len(df_es_top)} dominios .es")
df_es_top.head(20)
```
